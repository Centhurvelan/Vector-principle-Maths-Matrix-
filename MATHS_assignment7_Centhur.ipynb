```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)
A = np.array([
 [ 4, -2, 2, 1],
 [ 1, 1, 0, 1],
 [-2, 1, 3, -1],
 [ 1, 3, -1, 2]
], dtype=float)
A
```

1) Calculating the inverse matrix of A

```python
detA = np.linalg.det(A)
A_I = np.linalg.inv(A)
print("det(A) =", detA)
print("A_inverse =\n", A_I)
I_check = A @ A_I
print("\nA @ A_inverse :\n", I_check)
```

2. Calculating Eigenvalues and Eigenvectors

```python
eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues:\n", eigvals)
print("\nEigenvectors :\n", eigvecs)
idx = np.argsort(-np.abs(eigvals))
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]
print("\n Sorted Eigenvalues:\n", eigvals)
print("\n Corresponding Eigenvectors (columns):\n", eigvecs)
```

3. After having the Eigenvalues and Eigenvectors,

    - Constructing the eigen matrix P from eigenvectors: 

    - Constructing the diagonal matrix D from eigenvalues: 

    - Calculating the inverse of the eigen matrix:
  
    - 
    Verify if the diagonalized matrix `A_diagonalized` is identical to the original matrix `A` 


```python
P = eigvecs
D = np.diag(eigvals)
P_inv = np.linalg.inv(P)
A_diagonalized = P @ D @ P_inv
print("P:\n", P)
print("\nD:\n", D)
print("\nP^{-1}:\n", P_inv)
print("\nA_diagonalized:\n", A_diagonalized)
print("\nnp.allclose(A_diagonalized, A)? ->", np.allclose(A_diagonalized, A))  # Fixed: completed the function call
print("Max abs difference:", np.abs(A_diagonalized - A).max())
```

4)  Explain why this result (a diagonalized matrix is identical to the original matrix) is important in linear algebra (1 pts).

Diagonalization ($A = PDP^{-1}$) is crucial because it simplifies a complex matrix $A$ into a simple diagonal matrix $D$ by viewing it from the "right" perspective (the eigenvectors in $P$).This is vital because it makes complex operations, like calculating high powers ($A^{100}$), incredibly fast and easy. Instead of multiplying $A$ 100 times, you just take the power of $D$'s diagonal entries ($A^{100} = PD^{100}P^{-1}$), which is essential for analyzing dynamic systems and differential equations.

5)  Make a python code using SVD example to get the same A (AR)

  - Decide your own range for the sigma



```python
U, S, Vt = np.linalg.svd(A, full_matrices=True)
print("σ:", S)
def svd_reconstruct(U, S, Vt, r):
 Ur = U[:, :r]
 Sr = np.diag(S[:r])
 Vtr = Vt[:r, :]
 return Ur @ Sr @ Vtr
for r in range(1, 6): 
 A_r = svd_reconstruct(U, S, Vt, r)
 err = np.linalg.norm(A - A_r, ord='fro')
 print(f"r = {r}: ||A - A_r||_F = {err:.6e}")
A_4 = svd_reconstruct(U, S, Vt, 4)
print("\nnp.allclose(A, A_4)? ", np.allclose(A, A_4, atol=1e-10))
print("maximum abs difference (r=5):", np.abs(A - A_4).max())
```
